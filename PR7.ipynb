```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, train_test_split
from sklearn.metrics import f1_score, confusion_matrix, classification_report
import catboost
import time
```

```python
df = pd.read_csv('/content/drive/MyDrive/data science/Big Data/PR6/weather_forecast_data.csv')
print(df.head())
print('Емельянов')
```

```python
df.isna().sum()
```

```python
df['Rain'] = df['Rain'].replace({'no rain': 0, 'rain': 1})
```

```python
X = df.drop(columns=['Rain'])
y = df['Rain']
```

```python
X_train ,X_test,  y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
start_time = time.time()
cat_boost = catboost.CatBoostClassifier(random_state=43)
cat_boost.fit(X_train, y_train)
cat_res = cat_boost.predict(X_test)
end_time = time.time()
print(end_time - start_time)
print(f1_score(y_test, cat_res))
print(classification_report(y_test,  cat_res))
```

```python
start_time = time.time()
cl_tree = RandomForestClassifier(random_state=43)
cl_tree.fit(X_train, y_train)
end_time = time.time()
res_tree = cl_tree.predict(X_test)
print(end_time - start_time)
print(f1_score(y_test, res_tree))
print(confusion_matrix(y_test, res_tree))
print(classification_report(y_test, res_tree))
```
