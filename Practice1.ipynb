Задание 2

```python
import math

def calculate_area():
    figure = input("Введите название фигуры (треугольник, прямоугольник, круг): ").lower()

    if figure == 'треугольник':
        # Получаем длины сторон треугольника
        a = float(input("Введите длину первой стороны треугольника: "))
        b = float(input("Введите длину второй стороны треугольника: "))
        c = float(input("Введите длину третьей стороны треугольника: "))

        # Проверка условия существования треугольника
        if a + b > c and a + c > b and b + c > a:
            # Полупериметр
            p = (a + b + c) / 2
            # Площадь по формуле Герона
            area = math.sqrt(p * (p - a) * (p - b) * (p - c))
        else:
            return "Ошибка: такой треугольник не существует!"

    elif figure == 'прямоугольник':
        width = float(input("Введите ширину прямоугольника: "))
        height = float(input("Введите высоту прямоугольника: "))
        area = width * height

    elif figure == 'круг':
        radius = float(input("Введите радиус круга: "))
        area = math.pi * radius ** 2

    else:
        return "Ошибка: неизвестная фигура!"

    # Возвращаем результат в виде словаря
    return {figure: area}

# Вызываем функцию и выводим результат
result = calculate_area()
print(result)
```

Задание 3

```python
# Получаем ввод от пользователя
a = float(input("Введите первое число: "))
b = float(input("Введите второе число: "))
operation = input("Введите операцию (+, -, /, //, abs, pow, **): ")

# Выполняем соответствующую операцию
if operation == '+':
    result = a + b
elif operation == '-':
    result = a - b
elif operation == '/':
    if b != 0:
        result = a / b
    else:
        result = "Ошибка: деление на ноль!"
elif operation == '//':
    if b != 0:
        result = a // b
    else:
        result = "Ошибка: деление на ноль!"
elif operation == 'abs':
    result = abs(a), abs(b)
elif operation == 'pow' or operation == '**':
    result = a ** b
else:
    result = "Ошибка: неизвестная операция!"

# Выводим результат
print("Результат:", result, 'Емельянов')
```

```python
A = [1, 2, 3, 4, 2, 1, 3, 4, 5, 6, 5, 4, 3, 2]
B = ['a', 'b', 'c', 'c', 'c', 'b', 'a', 'c', 'a', 'a', 'b', 'c', 'b', 'a']
res = {}

for i in range(len(A)):
    if B[i] in res:
        res[B[i]] += A[i]
    else:
        res[B[i]] = A[i]

print(res, 'Емельянов')
```
