```python
# Проверка центральной предельной теоремы
n_samples = 300
sample_sizes = [10, 50, 100]
print('Емельянов')
for n in sample_sizes:
    sample_means = [df_insurance['charges'].sample(n).mean() for _ in range(n_samples)]
    plt.figure(figsize=(10, 6))
    sns.histplot(sample_means, kde=True)
    plt.title(f'Распределение средних значений для выборки размером {n}')
    plt.show()

    print(f'Для выборки размером {n}: Среднее = {np.mean(sample_means)}, Стандартное отклонение = {np.std(sample_means)}')
```

Центральная предельная теорема (ЦПТ) утверждает, что при достаточно большом размере выборки распределение выборочных средних будет стремиться к нормальному распределению, независимо от формы распределения исходной популяции.

Результаты:
Размер выборки n=10: Распределение выборочных средних имеет явные отклонения от нормальности, что подтверждает недостаточный размер выборки для применения ЦПТ.

Размер выборки n=50: Распределение выборочных средних демонстрирует более четкие характеристики нормального распределения, с симметричной формой и меньшим разбросом.

Размер выборки n=100: При размере выборки 100 распределение становится почти нормальным. Это подтверждает, что увеличение размера выборки улучшает соответствие нормальному распределению.

```python
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = sts.sem(data)
    margin_of_error = std_err * sts.t.ppf((1 + confidence) / 2, n - 1)
    return (mean - margin_of_error, mean + margin_of_error)


bmi_ci_95 = confidence_interval(df_insurance['bmi'], 0.95)
bmi_ci_99 = confidence_interval(df_insurance['bmi'], 0.99)


charges_ci_95 = confidence_interval(df_insurance['charges'], 0.95)
charges_ci_99 = confidence_interval(df_insurance['charges'], 0.99)
print('Емельянов')
print(f'95% доверительный интервал для BMI: {bmi_ci_95}')
print(f'99% доверительный интервал для BMI: {bmi_ci_99}')
print(f'95% доверительный интервал для Charges: {charges_ci_95}')
print(f'99% доверительный интервал для Charges: {charges_ci_99}')
```

```python


from scipy.stats import kstest, probplot

def check_normality(feature):
    ks_statistic, p_value = kstest(df_insurance[feature], 'norm', args=(df_insurance[feature].mean(), df_insurance[feature].std()))

    print(f"KS test for {feature}:")
    print(f"Statistic: {ks_statistic:.4f}, p-value: {p_value:.4f}")

    plt.figure(figsize=(6,6))
    probplot(df_insurance[feature], dist="norm", plot=plt)
    plt.title(f"Q-Q plot for {feature}")
    plt.show()

check_normality('bmi')
check_normality('charges')
```

```python

df_bmi = pd.read_csv('/content/bmi (1).csv')

northwest_bmi = df_bmi[df_bmi['region'] == 'northwest']['bmi']
southwest_bmi = df_bmi[df_bmi['region'] == 'southwest']['bmi']


from scipy.stats import shapiro

northwest_normality = shapiro(northwest_bmi)
southwest_normality = shapiro(southwest_bmi)

print(f'Нормальность для Northwest: p-value = {northwest_normality.pvalue}')
print(f'Нормальность для Southwest: p-value = {southwest_normality.pvalue}')


from scipy.stats import bartlett

bartlett_test = bartlett(northwest_bmi, southwest_bmi)
print(f'Bartlett test: p-value = {bartlett_test.pvalue}')


from scipy.stats import ttest_ind

t_test = ttest_ind(northwest_bmi, southwest_bmi)
print(f'T-критерий Стьюдента: p-value = {t_test.pvalue}')
print('Емельянов')
```

```python

observed_values = [97, 98, 109, 95, 97, 104]
expected_values = [100, 100, 100, 100, 100, 100]


chi2_test = sts.chisquare(observed_values, expected_values)
print(f'Хи-квадрат тест: p-value = {chi2_test.pvalue}')
```
